### To get any relevant financial data for any stock eg dividend,markrt cap etc

- be able to complare stock based on fundamentals

- see how this can be used to findin the hightest yield out of an index

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yf.Ticker('aapl').info
```

```python
tickers=['AAPL','msft','ibm']
```

```python
infos=[]

for i in tickers:
    infos.append(yf.Ticker(i).info)
```

```python
infos
```

```python
fundamentals= ['dividendYield','marketCap','beta','forwardPE']
df=pd.DataFrame(infos)
df=df.set_index('symbol')
df
```

```python
df[df.columns[df.columns.isin(fundamentals)]]
```

```python
plt.bar(df.index,df.forwardPE,color=('k','orange','cyan'))
plt.show()
```

```python
tickers=pd.read_html('https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')[1].Symbol.to_list()
tickers
```

```python
infos=[]

for i in tickers:
    infos.append(yf.Ticker(i).info)
```

```python
infos
```

```python
df2=pd.DataFrame(infos)
df2=df2.set_index('symbol')
df2
```

```python
df2['dividendYield'].nlargest(5) # instead of dividend yield other parameters can also be used like PE ratio etc
```

```python
plt.figure(figsize=(14,5))
plt.bar(df2['shortName'],df2['dividendYield'])
plt.xticks(rotation=90)
plt.show()
```
