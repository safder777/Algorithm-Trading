## The Strategy 

Check the past 12 month performance of all Nifty stocks
Buy the top performers and hold them for 1 month.

```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
stocks=pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[2].Symbol
stocks.head()
```

```python
stocks= stocks +'.NS' #yfincane need to add this to non US stocks
```

```python
stocks=stocks.to_list()
```

```python
df=yf.download(stocks,start='2012-01-01')['Close']
df
```

```python
ret_df=df.pct_change()
ret_df
```

```python
mtl_ret=(ret_df +1).resample('M').prod() #monthly return
mtl_ret
```

```python
mtl_12=mtl_ret.rolling(12).apply(np.prod).dropna()
mtl_12
```

```python
top_=mtl_12.loc['2012-12-31'].nlargest(5) #finding top 5 performers
top_
```

```python
top_.name #relevant date of the top performer , in above bajaj fin has more than doubled
```

```python
mtl_ret[top_.name:][1:2] #this gives the next month to the top performer
```

```python
relevant_ret=mtl_ret[top_.name:][1:2][top_.index]
relevant_ret
```

```python
relevant_ret.mean(axis=1)  #This is telling us that if we bought the top performer it would be 0.996842 of its value
```

```python
def top_performers(date):
    all_=mtl_12.loc[date]
    top=all_.nlargest(5)
    relevant_ret=mtl_ret[top.name:][1:2][top.index]
    return (relevant_ret).mean(axis=1).values[0]
```

```python
top_performers('2012-12-31')
```

```python
mom_ret=[]
for date in mtl_12.index[:-1]:
    mom_ret.append(top_performers(date)) 

# passing the monthly return and storing in mom_ret, skipping the last one
# becoz we dont know the future return.
```

```python
pd.Series(mom_ret)
```

```python
pd.Series(mom_ret).prod() #this is 11 times return
```

Comparing it with nifty

```python
nifty=yf.download('^NSEI',start='2012-01-01')['Close']
```

```python
(nifty.pct_change()+1).prod()
```
