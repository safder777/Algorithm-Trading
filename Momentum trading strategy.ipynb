## Our Strategy
- pick a stock universe
- In our case Dow Jones Industrial Average

## performance Evaluation 

- Check the performance of every single stock over last 12 month(t-12)

- Skip the most recent month
- So we have 11 past months(windows)

- But the winners and short the losers
- Hold this portfolio for 1 month

```python
import yfinance as yf
import pandas as pd
import datetime as dt
from dateutil.relativedelta import relativedelta
```

```python
# get ticker symbols for the DGI
tickers=pd.read_html('https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')[1].Symbol.to_list()
tickers
```

```python
# Get Prices fo DJI component

start=dt.datetime(2018,1,31)
end=dt.datetime(2020,1,31)

df=yf.download(tickers,start,end)['Adj Close']
df
```

```python
# Calculate monthly returns by cumulation daily returns

mtl_ret=df.pct_change().resample('M').agg(lambda x:(x+1).prod()-1)
mtl_ret
```

```python
# CAlculate retuns over the past 11 months-

import numpy as np
past_11=(mtl_ret+1).rolling(11).apply(np.prod)-1
past_11
```

```python
formation=dt.datetime(2019,12,31)

end_measurement=formation - relativedelta(months=1)
formation
```

```python
end_measurement
```

```python
ret_12=past_11.loc[end_measurement]
ret_12=ret_12.reset_index()
ret_12
```

```python
ret_12['quintile']=pd.qcut(ret_12.iloc[:,1],5,labels=False)
ret_12
```

```python
# ret_12=ret_12.rename(columns={'index': 'Symbols'}, inplace=True)
# winners
```

```python
winners=ret_12[ret_12.quintile==4]
losers=ret_12[ret_12.quintile==0]
```

```python
winnerret=mtl_ret.loc[formation + relativedelta(months=1),df.columns.isin(winners.index)]
loserret=mtl_ret.loc[formation + relativedelta(months=1),df.columns.isin(losers.index)]

# from dateutil.relativedelta import relativedelta

# # Assuming 'formation' is a datetime object
# next_month = formation + relativedelta(months=1)

# # Now you can use 'next_month' to index your DataFrame
# winnerret = mtl_ret.loc[next_month, df.columns.isin(winners.index)]
# loserret = mtl_ret.loc[next_month, df.columns.isin(losers.index)]
```
