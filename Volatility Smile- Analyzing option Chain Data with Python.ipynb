```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker=yf.Ticker('AAPL')
```

```python
ticker.options #we get weekly expiry
```

```python
expiration=ticker.options[6]
expiration
```

```python
option_chain=ticker.option_chain(expiration)
option_chain
```

```python
calls=option_chain.calls
calls #from here take OI value to filter which have less liquidity
```

```python
calls=calls[calls.openInterest>=600]
calls
```

```python
#to get volatility smile striek price on x axis and IV on yaxix
```

```python
plt.figure(figsize=(10,6))
plt.plot(calls['strike'],calls['impliedVolatility'])
plt.xlabel("strike price")
plt.ylabel("Implied volatility")
```
